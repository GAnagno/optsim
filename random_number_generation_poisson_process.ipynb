# Exponential random numbers
## Define a function to generate exponential random numbers

```python
def exponential_rng(lam=1.0):
    """ Generates exponential random numbers.
    
    Keywords:
        lam (float): the rate parameter, the inverse expectation of the distribution.
    
    Returns:
        exponential random numbers with rate lam
    """
    r = np.random.rand()
    return -np.log(r) / lam
```

```python
def exponential_pdf(x, lam=1.0):
    return lam * np.exp(-lam * x)
```

## Test the function

```python
exponential_rng()
```

## Generate draws and compare to theoretical density

```python
np.random.seed(1234)
R = 10000
lam = 2
l_exp = []
for r in range(R):
    exp = exponential_rng(lam=lam)
    l_exp.append(exp)
```

```python
x = np.arange(0, 10 / lam, 1e-4)
y = exponential_pdf(x, lam=lam)
plt.title('Exponential random numbers')
plt.hist(l_exp, density=True, bins=100, label = 'Draws', ec='w', lw=.1)
plt.plot(x, y, label = 'Density')
plt.legend()
plt.show()
```

# Homogeneous Poisson process
## Define a function to generate arrivals according to Poisson process

```python
def homogeneous_poisson_process(lam=4, T=1):
    t = 0
    l_t = []
    while True:
        r = np.random.rand()
        t = t - np.log(r) / lam
        if t > T:
            return l_t
        else:
            l_t.append(t)
```

```python
np.random.seed(1234)
homogeneous_poisson_process()
```

## Generate homogeneous Poisson process

```python
l_hpp = []
np.random.seed(1234)
for r in range(R):
    hpp = homogeneous_poisson_process()
    l_hpp.extend(hpp)
```

## Plot arrival times

```python
plt.hist(l_hpp, density=False, bins=10, ec='w', lw=.1)
plt.title('Homogeneous Poisson process')
plt.xlabel('Arrival time within Poisson process period')
plt.ylabel('Arrival count')
plt.show()
```

# Nonhomogeneous Poisson process
## Define a function to generate arrivals according to non-homogeneous Poisson process

```python
def non_homogeneous_poisson_process(lam_t=4, lam_max=8, T=10):
    t = 0
    l_t = []
    while True:
        r = np.random.rand()
        t = t - np.log(r) / lam_max
        if t > T:
            return l_t
        else:
            s = np.random.rand()
            lam = lambda t: lam_t * np.sin(t) + lam_t
            if r <= lam(t) / lam_max:
                l_t.append(t)
```

```python
def lambda_t(x, lam_t=4.0):
    return lam_t * np.sin(x) + lam_t
```

## Generate nonhomogeneous poisson processes

```python
l_nhpp = []
R = 1000
np.random.seed(1234)
for r in range(R):
    nhpp = non_homogeneous_poisson_process()
    l_nhpp.extend(nhpp)
```

## Plot arrival times

```python
x = np.arange(0, 10, 1e-2)
y = lambda_t(x)
plt.title('Nonhomogeneous Poisson process')
plt.xlabel('Arrival time within Poisson process period')
plt.ylabel('Arrival count')
plt.hist(l_nhpp,
         density=False,
         bins=np.linspace(0, 10, 11),
         ec='w',
         lw=.1,
         label='Generated arrivals')
plt.plot(x, R * y, label='$R * lambda(t)$')
plt.legend()
plt.show()
```
